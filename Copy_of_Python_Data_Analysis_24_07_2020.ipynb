```
# Importing the pandas library
# 
import pandas as pd

# Importing the numpy library
#
import numpy as np
```

```
#now we read the data from the CSV file and 
#create the dataframe to be used
df = pd.read_csv('/content/Autolib_dataset (2).csv')

#previewing the first 10 rows of our dataframe
df.head(10)
```

```
#accessing the content of our dataset
df.info()
```

```
#checking sum of all missing values
#df.isnull().sum().sum()

all_rows = df.shape[0]
missing_num = all_rows - df.count() 
missing_num
```

```
#the number of non-missing value
df.count()
```

```
#combining the year,date,and day into a date time series
df['Year'] = pd.to_datetime(df[['year', 'month', 'day']])
```

```
#previwing our dataframe
df.head()
```

```
#now we drop irrelevent columns
df.drop(['Displayed comment', 'Geo point', 'Scheduled at', 'year',	'month'], axis=1, inplace=True)
```

```
#previewing the dataframe now
df.head()
```

```
#Identify the most popular hour of the day 
#for picking up a shared electric car (Bluecar) 
#in the city of Paris over the month of April 2018.
#
df.groupby(['hour', 'City'])['Bluecar counter'].sum().sort_values(ascending=False)
```

```
#df.head(10)

grp = df.groupby('Address')[['Bluecar counter', 'Utilib counter', 'Utilib 1.4 counter']] 
grp.sum()
```

```
#to join successive (in time) measures/counters for a given station,
# as the difference will tell whether a car was picked up, returned, or nothing happened.
dat = df.groupby(['Address','hour'], 
                  as_index=False)['Bluecar counter', 'Utilib counter', 'Utilib 1.4 counter'].agg(lambda x: list(x))
dat         
```

```
#changing the column elements to int from string in dat dataframe
dat1 = dat[['Bluecar counter',	'Utilib counter',	'Utilib 1.4 counter']] 
dat1['Bluecar counter'] = df['Bluecar counter'].astype(int)
dat1['Utilib counter'] = df['Utilib counter'].astype(int)
dat1['Utilib 1.4 counter'] = df['Utilib 1.4 counter'].astype(int)
dat1
#finding the diff for the columns to calculate usage
dat1.diff()
```

```
#joining the two new dataframes
df_4 = pd.merge(dat, dat1, right_index=True, left_index=True)
df_4
```
